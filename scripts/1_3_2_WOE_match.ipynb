In this notebook we assign to testing sample cateogories based on training sample their intervals.

```python
import pandas as pd
import numpy as np
import pickle
```

###  Create df testing with matched with new categories

```python
df_cont = pd.read_csv('../data/output/categorical_dfs/woe_auxiliary/testing_sample_cleaned_continuous.csv')
df_categ = pd.read_csv('../data/output/categorical_dfs/testing_sample_cleaned_categorical.csv')
columns_final_df = pd.read_csv('../data/output/categorical_dfs/columns_final_woe.csv')
```

```python
# change columns_final_df to list
columns_final_df = columns_final_df['0'].tolist()

# delete sufix '_woe' from columns_final_df if the name is different than 'target'
columns_final_df_without_suffix = [x[:-4] if x.endswith('_woe') and x != 'target' else x for x in columns_final_df]
columns_final_df_without_suffix
```

```python
df_cont.columns
```

```python
df_cont
```

```python
# read bin_edges
bin_edges = pd.read_csv('../data/output/categorical_dfs/woe_auxiliary/bin_edges.csv')
```

```python
with open('../data/output/categorical_dfs/woe_auxiliary/bin_edges.pkl', 'rb') as f:
    bin_edges = pickle.load(f)
```

```python
bin_edges
```

```python
# rename value in column_name in bin_edges
bin_edges.loc[bin_edges['column_name'] == 'Var8', 'column_name'] = 'Var8_q'
bin_edges.loc[bin_edges['column_name'] == 'Var17', 'column_name'] = 'Var17_1'
bin_edges.loc[bin_edges['column_name'] == 'Var25', 'column_name'] = 'Var25_q'
bin_edges.loc[bin_edges['column_name'] == 'Var26', 'column_name'] = 'Var26_q'
bin_edges.loc[bin_edges['column_name'] == 'working_months', 'column_name'] = 'working_months_1'
```

```python
bin_edges
```

```python
columns_final_df_without_suffix
```

```python
# take the common part of the two lists
common_columns = list(set(bin_edges['column_name']).intersection(set(columns_final_df_without_suffix)))
common_columns
```

```python
df_cont.columns
```

```python
df_cont.rename(columns = {'Var8':'Var8_q'}, inplace = True)
df_cont.rename(columns = {'Var17':'Var17_1'}, inplace = True)
df_cont.rename(columns = {'Var25':'Var25_q'}, inplace = True)
df_cont.rename(columns = {'Var26':'Var26_q'}, inplace = True)
df_cont.rename(columns = {'Var26':'Var26_q'}, inplace = True)
df_cont.rename(columns = {'working_months':'working_months_1'}, inplace = True)
```

```python
df_cont1 = df_cont[common_columns]
```

### 

Variables 8, 25, 26 if NA than 10

```python
df_cont1
```

```python
bin_edges
```

```python
# change the values of Var 8 from 999999999 to NA
df_cont1['Var8_q'] = df_cont1['Var8_q'].replace(999999999, np.nan)
df_cont1['Var25_q'] = df_cont1['Var25_q'].replace(999999999, np.nan)
df_cont1['Var26_q'] = df_cont1['Var26_q'].replace(999999999, np.nan)
```

```python
df_cont1
```

```python
df_cont1.isna().sum()
```

```python
# Apply the category from 0 to 9 to each column in df_cont1 basing on bin_edges
column = 'Var4'
```

```python
# Function to apply bin edges to appropriate columns
def apply_bin_edges(row):
    column_name = row['column_name']
    bin_edges = row['bin_edges']
    return pd.cut(df_cont1[column_name], bins=bin_edges, labels=False, include_lowest=True)
```

```python
# select df bin_edges where column name is in common_columns
bin_edges_final = bin_edges[bin_edges['column_name'].isin(common_columns)]
bin_edges_final
```

```python
# Apply bin edges to appropriate columns
for _, row in bin_edges_final.iterrows():
    df_cont1[row['column_name'] + '_binned'] = apply_bin_edges(row)

for _, row in bin_edges_final.iterrows():
    df_cont1[row['column_name'] + '_binned'] = apply_bin_edges(row)
```

```python
df_cont1
```

```python
df_cont1.isna().sum()
```

Fill NAs with category10

```python
# Fill Var25_q_binned, Var26_q_binned and var8_binned NAs with 10
df_cont1['Var25_q_binned'] = df_cont1['Var25_q_binned'].fillna(10)
df_cont1['Var26_q_binned'] = df_cont1['Var26_q_binned'].fillna(10)
df_cont1['Var8_q_binned'] = df_cont1['Var8_q_binned'].fillna(10)
```

```python
df_cont1
```

```python
# select all columns that end with _binned
df_cont1_binned = df_cont1.filter(regex='_binned')
df_cont1_binned
```

```python
# select columns that are in columns_final_df_without_suffix but not in common_columns
remaining_columns = list(set(columns_final_df_without_suffix).difference(set(common_columns)))
remaining_columns
```

```python
df_categorical = df_categ[remaining_columns]
```

```python
df_categorical.shape
```

```python
df_cont1_binned.shape
```

```python
# merge df_cont1_binned and df_categorical
df_final = pd.concat([df_cont1_binned, df_categorical], axis=1)
df_final
```

```python
# remove sufixes _binned from columns
df_final.columns = df_final.columns.str.replace('_binned', '')
df_final
```

Now we have testing dataset where countinuous variable have categories basing on interval from development df
